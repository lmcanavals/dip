```python
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
from scipy.ndimage.morphology import binary_erosion, binary_dilation
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = rgb2gray(imread('imagenes/huella_1.jpg'))
thres = threshold_otsu(im)
im = (im > thres).astype(np.uint8)

plt.imshow(im, cmap='gray'), plt.axis('off')
```

```python
im1 = 1 - im
im2 = binary_erosion(im1, structure=np.ones((3, 3)))
im3 = binary_dilation(im2, structure=np.ones((3, 3)))
im4 = binary_dilation(im3, structure=np.ones((2, 2)))
im5 = binary_erosion(im4, structure=np.ones((2, 2)))
im6 = 1 - im5
```

```python
plt.figure(figsize=(30, 20))
plt.gray()
plt.subplots_adjust(left=0, right=1, bottom=0, top=0.95, hspace=0.05, wspace=0.05)

plt.subplot(231), plt.imshow(im1), plt.axis('off')
plt.title('Otsu', size=20)

plt.subplot(232), plt.imshow(im2), plt.axis('off')
plt.title('erosion', size=20)

plt.subplot(233), plt.imshow(im3), plt.axis('off')
plt.title('apertura', size=20)

plt.subplot(234), plt.imshow(im4), plt.axis('off')
plt.title('apertura + dilasion', size=20)

plt.subplot(235), plt.imshow(im5), plt.axis('off')
plt.title('apertura + cierre', size=20)

plt.subplot(236), plt.imshow(im6), plt.axis('off')
plt.title('inv', size=20)
```
