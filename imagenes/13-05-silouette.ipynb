```python
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
from scipy.ndimage.morphology import binary_erosion, binary_dilation
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = rgb2gray(imread('imagenes/siluetas-3.jpg'))
thres = threshold_otsu(im)
im = (im > thres).astype(np.uint8)

plt.imshow(im, cmap='gray'), plt.axis('off')
```

```python
im1 = 1 - im
im2 = binary_erosion(im1, structure=np.ones((3, 3)))
im3 = im1 - im2

plt.subplots(figsize=(10, 10))
plt.imshow(im3, cmap='gray'), plt.axis('off')
```
