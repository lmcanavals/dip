# RGB to Gray

$color = 0.2989 * r + 0.5870 * g + 0.1140 * b$

```python
def rgb2grey1(im):
  return (im[:,:,0] + im[:,:,1] + im[:,:,2]) // 3
def rgb2grey2(im):
  return np.floor(0.2989 * im[:,:,0] + 0.587 * im[:,:,1] + 0.114 * im[:,:,2])
```

```python
plt.imshow(rgb2grey1(im), cmap="Greys_r")
```

```python
plt.imshow(rgb2grey2(im), cmap="Greys_r")
```
