```python
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
from scipy.ndimage.morphology import binary_erosion, binary_dilation
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = rgb2gray(imread('imagenes/erosion1.png'))
thres = threshold_otsu(im)
im = (im > thres).astype(np.uint8)

eroded1 = binary_erosion(im, structure=np.ones((11, 11)))
eroded2 = binary_erosion(im, structure=np.ones((15, 15)))
eroded3 = binary_erosion(im, structure=np.ones((45, 45)))
```

```python
plt.figure(figsize=(20, 20))
plt.gray()
plt.subplots_adjust(left=0, right=1, bottom=0, top=0.95, hspace=0.05, wspace=0.05)

plt.subplot(221), plt.imshow(im), plt.axis('off')
plt.title('thresholded original image (otsu)', size=20)

plt.subplot(222), plt.imshow(eroded1), plt.axis('off')
plt.title('eroded image (11x11 square)', size=20)

plt.subplot(223), plt.imshow(eroded2), plt.axis('off')
plt.title('eroded image (15x15 square)', size=20)

plt.subplot(224), plt.imshow(eroded3), plt.axis('off')
plt.title('eroded image (45x45 square)', size=20)
```
