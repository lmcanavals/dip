```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
base = "https://raw.githubusercontent.com/lmcanavals/dip/main/imagenes/"
```

```python
img1 = "eminem.jpg"
url = f"{base}{img1}"
!curl -O $url
```

```python
img2 = "gradient.png"
url = f"{base}{img2}"
!curl -O $urli
```

```python
def show(im, size=8):
  _, ax = plt.subplots(figsize=(size, size))
  ax.imshow(im, interpolation="nearest")
  plt.tight_layout()
  plt.show()
```

```python
def loadImg(img):
  im = cv2.imread(img)
  return im[:,:,::-1]
```

```python
eminem = loadImg(img1)
show(eminem)
```

```python
gradient = loadImg(img2)
show(gradient)
```

## 2 Images operations

## Addition

```python
eminem = eminem / 255
gradient = gradient / 255
```

```python
fig = plt.figure(figsize=(12, 6))
for i in range(12):
  alpha = (i + 0.5) / 12
  a = fig.add_subplot(3, 4, i + 1)
  plt.axis('off')
  aux = alpha * gradient + (1 - alpha) * eminem
  plt.imshow(aux)
  plt.text(0, 0, f"α={alpha:.3f}")
plt.show()
```

## Addition 2

```python
img3 = "zuckerberg.jpg"
url = f"{base}{img3}"
!curl -O $url
```

```python
zuckerberg = loadImg(img3)
```

```python
zuckerberg = zuckerberg / 255
```

```python
aux = (eminem + zuckerberg) / 2
show(aux)
```

## Difference


```python
fig = plt.figure(figsize=(20, 20))
a = fig.add_subplot(3, 4, 1)
plt.axis('off')
aux = abs(zuckerberg - eminem)
plt.imshow(aux)
plt.text(0, 0, f"eminem - zuckerberg")
a = fig.add_subplot(3, 4, 2)
plt.axis('off')
aux = abs(eminem - zuckerberg)
plt.imshow(aux)
plt.text(0, 0, f"zuckerberg - eminem")

plt.show()
```
